### Step 5: Handle Missing Values (Data Cleaning)

- Handle missing values in Age column


```python
df["Age"].isnull().sum()

age = df["Age"].mean()

df["Age"] = df["Age"].fillna(age)
```

- Check if any missing value is remaining

```python
df["Age"].isnull().sum()  
```

- Handle missing values in Salary

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].median())
```

```python
df["Salary"].isnull().sum() 
```

- Handle missing values in Experience_Years 

```python

df["Experience_Years"] = df["Experience_Years"].fillna(df["Experience_Years"].median())
```

```python
df["Experience_Years"].isnull().sum()
```

- Check all missing values handle or not

```python
df.isnull().sum()
```

- Missing values were handled using mean, median, and default values based on the column type.

### Step 6: Data Filtering

```python
df[df["Salary"]>80000]

```

### Step 7 Data Sorting

### Step 8: GroupBy Analysis

Average salary per department

```python
df.groupby("Department")["Salary"].mean()
```
